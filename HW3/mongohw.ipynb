```python
# Aggregation
#1
result = db.movies.aggregate([
    {
        '$group': {
            '_id': None, 
            'avgAwards': {'$avg': '$awards.wins'}
        }
    }
])

for doc in result:
    print(doc['avgAwards'])
```

```python
#2
result = db.movies.aggregate([
    {
        '$group': {
            '_id': None, 
            'maxAwards': {'$max': '$awards.wins'},
            'title': {'$first': '$title'},
            'awards': {'$first': '$awards'}
        }
    }
])

for doc in result:
    print(doc['maxAwards'])
```

```python
#3
result = db.movies.aggregate([
    {
        '$group': {
            '_id': None, 
            'count': {'$sum': 1}
        }
    }
])

for doc in result:
    print(doc['count'])
```

```python
#4
result = db.comments.aggregate([
    {
        '$group': {
            '_id': '$name',
            'count': {'$sum': 1}
        }
    },
    {
        '$project': {
            'name': '$_id',
            '_id': 0,
            'count': 1
        }
    }
])

user_count = len(db.comments.distinct('name'))

print(user_count)

# If you want to see all of the name and their counts
# for doc in result:
#     print(doc['name'], doc['count'])
```

```python
#5
result = db.comments.aggregate([
    {
        "$group": {
            "_id": "$name",
            "count": { "$sum": 1 }
        }
    },
    {
        "$limit": 5
    }
])

for doc in result:
    print(doc)
```
